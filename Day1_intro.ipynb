# 📌 ETF Total Return Analysis – Project Introduction

## 📌 Background

Many investors evaluate ETFs primarily based on price performance.  
However, this approach overlooks a critical component of total return: dividends.

Dividend-paying ETFs can generate a significant portion of their returns through consistent income, which may not be reflected in price movements alone.

This project was initiated to analyze ETF performance more accurately by incorporating both price changes and dividend effects.

---

## ❗ Problem Statement

Price-based analysis alone does not fully capture the actual performance of ETFs.

In particular, dividend-focused ETFs such as SCHD and JEPQ may appear to underperform when only price returns are considered, despite generating substantial income through dividends.

---

## 🎯 Objective

The objective of this project is to evaluate ETF performance from multiple perspectives:

- Price return (capital appreciation)
- Dividend yield (income component)
- Total return (including dividend reinvestment)

---

## 📊 ETF Selection

The following ETFs were selected to represent distinct investment strategies:

- QQQ: Growth-oriented ETF focusing on technology and innovation  
- SCHD: Dividend-focused ETF with stable income  
- JEPQ: Income-focused ETF utilizing option premium strategies  

This selection enables a comparative analysis of different return structures.

---

## 🔍 Expected Outcome

This project aims to demonstrate that:

- Dividend reinvestment plays a significant role in long-term investment performance  
- Compared to QQQ, JEPQ tends to show lower price growth during bull markets but provides relatively better downside protection during market downturns  
- Due to its income-generating structure, including option premium and dividends, JEPQ can achieve competitive total returns despite lower price appreciation  
- SCHD is expected to provide more stable returns with moderate growth and consistent dividend income
